```python
#Dependencies
import pandas as pd
```

```python
#READ IN CSVs
```

```python
#Set year
year = "2020"

#list of months
months = ["01", "02", "03", "04", "05", "06", "07"]

#Create empty df
df_20 = pd.DataFrame()

#Loop through months to sample and append to df
for month in months:
    temp_df = pd.read_csv(f'../Resources/{year}_rides/{year}{month}-citibike-tripdata.csv')
    temp_df = temp_df.sample(frac=.1, random_state = 88)
    df_20 = df_20.append(temp_df)
    
    print(month)

```
